```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('lsd_math_score_data.csv')
data
```

```python
time = data[['Time_Delay_in_Minutes']] # when wrap inside [[]] it means we save the object to a dataframe instead of series
LSD = data[['LSD_ppm']]
score = data[['Avg_Math_Test_Score']]
```

```python
# tell Jupyter to export the chart along with the notebook when download it
%matplotlib inline

# Style the plot graph
plt.title('Tissue concentration of LSD over time', fontsize=17)
plt.xlabel('Time in Minutes', fontsize=14)
plt.ylabel('Tissue LSD ppm', fontsize=14)
plt.text(x=0, y=-0.5, s='Wagner et al. (1968)', fontsize=12)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(1, 7)
plt.xlim(0, 500)

plt.style.use('classic')

# Plot the Graph data
plt.plot(time, LSD, c='#e74c3c', linewidth=3)
plt.show()
```

```python
regr = LinearRegression()
regr.fit(LSD, score)
print('Theta1: ', regr.coef_[0][0])
print('Intercept: ', regr.intercept_[0])
print('R-Squeare: ', regr.score(LSD, score))
predicted_score = regr.predict(LSD)
```

```python
%matplotlib inline

plt.title('Arithmetic vs LSD-25', fontsize=14)
plt.xlabel('Tissue LSD ppm', fontsize=14)
plt.ylabel('Performance Score', fontsize=14)
plt.ylim(25, 85)
plt.xlim(1, 6.5)
plt.style.use('fivethirtyeight')
plt.scatter(LSD, score, c='blue', alpha=0.5, s=100)

plt.plot(LSD, predicted_score, c='red', lw=3)
plt.show()
```
